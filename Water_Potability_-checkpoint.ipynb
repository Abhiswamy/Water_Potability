## Data Preprocessing

```python
X=df.drop('Potability', axis=1)
y=df.Potability
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
X = pd.DataFrame(X)
```

```python
oversample = SMOTE()
X, y = oversample.fit_resample(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.28, random_state=101)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train))
X_test=pd.DataFrame(scaler.transform(X_test))
```

## Modelling

```python
models = pd.DataFrame(columns=["Model", "Train_Score", "Test_Score"])
```

```python
model_list = [("Random Forest", RandomForestClassifier(n_estimators=1000)),
              ("LightGBM", LGBMClassifier(n_estimators=1000)),
              ("Support Vector Machines", SVC()),
              ("XGBoost", XGBClassifier(n_estimators=1000)),
              ("Gaussian Naive Bayes", GaussianNB()),
              ("KNN", KNeighborsClassifier()),
              ("Decision Tree", DecisionTreeClassifier()),
              ("Bagging Classifier", BaggingClassifier(base_estimator=RandomForestClassifier(), random_state=42))]
```

```python
results_test = []
names_test = []
finalResults_test = []

results_train = []
names_train = []
finalResults_train = []

for name,model in model_list:
    model.fit(X_train, y_train)
    
    # Predictions on test data
    model_results_test = model.predict(X_test)
    score_test = precision_score(y_test, model_results_test)
    results_test.append(score_test)
    names_test.append(name)
    finalResults_test.append((name,score_test))
    
    # Predictions on training data
    model_results_train = model.predict(X_train)
    score_train = precision_score(y_train, model_results_train)
    results_test.append(score_train)
    names_train.append(name)
    finalResults_train.append((name,score_train))
    
    new_row={'Model':name, 'Train_Score': score_train, 'Test_Score': score_test}
    models=models.append(new_row, ignore_index=True)
```

```python
models.sort_values(by="Test_Score", ascending=False)
```

```python
models = pd.DataFrame(columns=["Model", "Accuracy Score"])
```

```python
model_list = [("Random Forest", RandomForestClassifier(n_estimators=1000)),
              ("LightGBM", LGBMClassifier(n_estimators=1000)),
              ("Support Vector Machines", SVC()),
              ("XGBoost", XGBClassifier(n_estimators=1000)),
              ("Gaussian Naive Bayes", GaussianNB()),
              ("KNN", KNeighborsClassifier()),
              ("Decision Tree", DecisionTreeClassifier()),
              ("Bagging Classifier", BaggingClassifier(base_estimator=RandomForestClassifier(), random_state=42))]
```

```python
for name, clf in model_list:
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    score = precision_score(y_test, predictions)
    
    new_row = {"Model": name, "Accuracy Score": score}
    models = models.append(new_row, ignore_index=True)
```

```python
models.sort_values(by="Accuracy Score", ascending=False)
```

```python
clf = RandomForestClassifier(n_estimators=1000)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
acc = precision_score(y_test, y_pred) * 100
print("Accuracy =", acc)
```

## Hyperparameter Tuning

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
 
g_svc = GridSearchCV(SVC(), param_grid, refit = True)
g_svc.fit(X_train, y_train)
```

```python
svc_predict=g_svc.predict(X_test)
sv_acc=accuracy_score(y_test,svc_predict)
print(f'Accuracy = {sv_acc}')
print(classification_report(y_test, svc_predict))
```

```python
estimator = XGBClassifier()
parameters = {
    'max_depth': [2,5,8],
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05]
}
xgb_grid = GridSearchCV(
    estimator=estimator,
    param_grid=parameters,
    scoring = 'roc_auc',
)

xgb_grid.fit(X_train, y_train)
```

```python
xgb_predict=xgb_grid.predict(X_test)
xgb_acc=accuracy_score(y_test,xgb_predict)
print(f'Accuracy = {xgb_acc}')
print(classification_report(y_test, xgb_predict))
```

```python
rf_param = {'bootstrap': [True, False],
 'max_depth': [30, 40, 50],
 'min_samples_leaf': [3,5],
 'min_samples_split': [5],
 'max_features':['auto','sqrt','log2'],
 'n_estimators': [500]}
rf_grid = GridSearchCV(RandomForestClassifier(), param_grid = rf_param, cv = 3)
rf_grid.fit(X_train, y_train)
```

```python
rf_predict=rf_grid.predict(X_test)
rf_acc=accuracy_score(y_test,rf_predict)
print(f'Accuracy = {rf_acc}')
print(classification_report(y_test, rf_predict))
```
